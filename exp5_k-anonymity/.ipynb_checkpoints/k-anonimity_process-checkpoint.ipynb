# k-anonimity实验

```python
import pandas as pd
import matplotlib.pyplot as plt

file_name = "adult_with_pii.csv"
adult = pd.read_csv(file_name)
adult.head()
```

```python
adult_data = adult.copy().drop(columns=['Name', 'SSN'])
adult_data.to_csv('adult_data_without_name_and_ssn.csv')
```

```python
adult_pii = adult[['Name', 'SSN', 'DOB', 'Zip']]
adult_pii.to_csv('adult_data_with_name_ssn_dob_and_zip.csv')
```

## 链接攻击
通过去识别化的数据确定一个人

```python
# 使用生日和邮政编码去匹配特定人员，我们可以获得他的信息
karries_row = adult_pii[adult_pii['Name'] == 'Karrie Trusslove']
pd.merge(karries_row, adult_data, left_on=['DOB', 'Zip'], right_on=['DOB', 'Zip'])
```

```python
# 同理对wash_gives进行链接攻击
wash_gives_row = adult_pii[adult_pii['Name'] == 'Wash Gives']
pd.merge(wash_gives_row, adult_data, left_on=['DOB', 'Zip'], right_on=['DOB', 'Zip'])
```

```python
# 只使用生日进行攻击

pd.merge(karries_row, adult_data, left_on=['DOB'], right_on=['DOB', ])
```

```python
# 只使用邮政编码进行攻击
pd.merge(karries_row, adult_data, left_on=['Zip', ], right_on=['Zip', ])
```

```python
adult_pii['DOB'].value_counts().hist()
plt.xlabel('Number of Dates of Birth')
plt.ylabel('Number of Occurrences')
plt.show()
```

```python
adult_pii['Zip'].value_counts().hist()
plt.xlabel('Number of ZIP Codes')
plt.ylabel('Number of Occurrences');
```
